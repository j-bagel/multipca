## Multi-way PCA Examples

This notebook demonstrates the use of Multiway Principal Component Analysis (MultiPCA) with confidence intervals on two different datasets:

1. **Citibike Data**: Bike sharing patterns in New York City
   - Dimensions: (24 hours × 35 stations × 522 days)
   - Analysis of daily usage patterns across stations

2. **World Bank Data**: 9 economic indicators over 19 years, across 160 countries
   - Multiple economic metrics over time
   - Analysis of trends and patterns between indicators

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as st

import multipca
```

## Citibike Analysis

```python
array_citi = np.load('citibike_data.npy')
array_citi.shape
```

```python
results_citi = multipca.mpca_CI(array_citi, 8, stability=True)
```

### Daily Pattern Components (24h)

```python
ax = multipca.CI_band_plot(results_citi, rank=1, mode=1, confidence=0.95)
plt.show()
```

```python
ax = multipca.CI_band_plot(results_citi, rank=2, mode=1, confidence=0.95)
plt.show()
```

```python
ax = multipca.CI_band_plot(results_citi, rank=3, mode=1, confidence=0.95)
plt.show()
```

## World Bank Data Analysis

```python
array_wb = np.load('worldbank_data.npy')
array_wb.shape
```

```python
results_wb = multipca.mpca_CI(array_wb, 5, stability=True)
```

### Temporal Components (2000-2018)

```python
ax = multipca.CI_band_plot(results_wb, rank=1, mode=1)
ax.set_xticks([0,8,18], [2000,2008,2018])
plt.show()
```

```python
ax = multipca.CI_band_plot(results_wb, rank=2, mode=1, flip = True)
ax.set_xticks([0,8,18], [2000,2008,2018])
plt.show()
```

```python
ax = multipca.CI_band_plot(results_wb, rank=3, mode=1)
ax.set_xticks([0,8,18], [2000,2008,2018])
plt.show()
```

### Economic Indicator Analysis

```python
labels = ["GDP","import", "export", "CPI", "life span", "CO2","birth","urban","tourism"]
```

```python
ax = multipca.CI_plot(results_wb, rank=1, mode=2, labels = labels, y_as_value=False, flip=True)
plt.show()
```

```python
ax = multipca.CI_plot(results_wb, rank=2, mode=2, labels = labels, y_as_value=False, flip=True)
plt.show()
```

```python
ax = multipca.CI_plot(results_wb, rank=3, mode=2, labels = labels, y_as_value=False, flip=True)
plt.show()
```

#### A loading plot example

```python
ax = multipca.loading_plot(result=results_wb, rank1=2, rank2=3, mode=2, labels=labels)
```
